```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python

X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')
```

```python
X_train_num = X_train.copy()
X_train_num = X_train_num.select_dtypes(include=['int64', 'float64'])
```

```python
X_train_num['OECD'].corr(X_train_num['ST331'], method='spearman')
```

```python
y_train = y_train["MathScore"].copy()
```

```python
X_train = X_train.dropna(axis=1, how='all')
```

```python
X_small, X_test_small, y_small, y_test_small = train_test_split(
    X_train, y_train,
    train_size=0.8,
    test_size=0.2,
    random_state=42
)
```

```python
X_small = X_small.dropna(axis=1, how='all')
```

```python

cat_columns = X_small.select_dtypes(include='object').columns.tolist()
num_columns = X_small.select_dtypes(include=['int64','float64']).columns.tolist()
```

```python

num_transformer = Pipeline([
('imputer', SimpleImputer(strategy='mean')),
('scaler', StandardScaler())
])

cat_transformer = Pipeline([
('imputer', SimpleImputer(strategy='most_frequent')),
('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer(
transformers=[
('num', num_transformer, num_columns),
('cat', cat_transformer, cat_columns)
],
remainder='drop'
)

```

```python

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
pipeline.fit(X_small, y_small)
r2_score = pipeline.score(X_test_small, y_test_small)
print("R^2 score:", r2_score)
```
