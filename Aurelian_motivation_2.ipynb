```python
import pandas as pd
pd.set_option('display.max_rows', None)
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
from sklearn.feature_selection import SelectFromModel
```

```python
# Importation
X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')
df = pd.merge(X_train, y_train)
```

```python
sns.countplot(x='ST301', data=df, palette='Set2')
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Distribution of ST301")
plt.show()
```

```python
sns.scatterplot(x='ST322', y='ST301', data=df, alpha=0.5)
plt.xlabel("Disturbation")
plt.ylabel("Motivation")
plt.title("Scatter plot")
plt.show()
```

```python
plt.hexbin(df['ST322'], df['ST301'], gridsize=30, cmap='viridis')
plt.colorbar(label='Counts')
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.title("Hexbin plot of quasi-continuous variables")
plt.show()
```

```python
sns.kdeplot(x='ST322', y='ST301', data=df, fill=True, cmap="mako", thresh=0.05)
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.title("2D density")
plt.show()
```
